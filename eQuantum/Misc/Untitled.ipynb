```python
import numpy as np
from pyquil.api import WavefunctionSimulator

from grove.pyqaoa.maxcut_qaoa import maxcut_qaoa
```

```python
steps = 5
square_ring = [(0,1),(1,2),(2,3),(3,4),(4,0),(3,0)]

inst = maxcut_qaoa(square_ring, steps=steps)
opt_betas, opt_gammas = inst.get_angles()
```

```python
t = np.hstack((opt_betas, opt_gammas))
param_prog = inst.get_parameterized_program()
prog = param_prog(t)
wf = WavefunctionSimulator().wavefunction(prog)
wf = wf.amplitudes
```

```python
from qiskit.visualization import plot_histogram
```

```python
res = {inst.states[state_index]: (np.conj(wf[state_index]) * wf[state_index]) for state_index in range(inst.nstates)}
plot_histogram(res, figsize = (20,10))
```
